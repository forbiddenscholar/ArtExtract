```python
published_images = published_images[['uuid', 'iiifurl', 'depictstmsobjectid', 'width', 'height']]
objects = objects[['objectid', 'title', 'attribution', 'classification']]
```

```python
## merging the dataset on basis of object id
merged_df = pd.merge(objects, published_images, left_on='objectid', right_on='depictstmsobjectid', how='inner')
## we will use only paintings (ignoring photograph, portfolio etc..)
merged_df = merged_df[merged_df['classification'] == 'Painting']
merged_df
```

```python
## creating the subset of the dataset
subset_df = merged_df.sample(500)
```

```python
image_urls = subset_df['iiifurl'].tolist()
```

```python
import os
import requests
from concurrent.futures import ThreadPoolExecutor

download_dir = "downloaded_images"
os.makedirs(download_dir, exist_ok=True)

def download_image(url):

    image_id = url.split("/")[-1]

    image_url = f"{url}/full/full/0/default.jpg"
    filename = f"{image_id}.jpg"
    filepath = os.path.join(download_dir, filename)

    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        r = requests.get(image_url, headers=headers, stream=True, timeout=20)
        r.raise_for_status()

        with open(filepath, "wb") as f:
            for chunk in r.iter_content(8192):
                if chunk:
                    f.write(chunk)

        print("Downloaded:", filename)

    except Exception as e:
        print("Failed:", image_id, e)


with ThreadPoolExecutor(max_workers=15) as executor:
    executor.map(download_image, image_urls)
```
